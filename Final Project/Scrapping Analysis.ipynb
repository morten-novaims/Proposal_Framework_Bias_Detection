## Analysis of the log scrapping file

Read the file

```python
with open('log.txt', 'r') as f:
    content = f.readlines()
content = [x.strip() for x in content if "Got" in x] 
```

Parse the file

```python
from datetime import datetime
log_result=[]
for line in content:
    parsed = line.split(' ')
    date = datetime.strptime(parsed[0]+' '+parsed[1], '%Y-%m-%d %H:%M:%S.%f:')
    log_result.append((date, int(parsed[3]))) 
```

Convert to dataframe

```python
import pandas as pd
log_df= pd.DataFrame(log_result)
log_df.columns = ['time', 'log_count']
```

Create the plot and save it

```python
from matplotlib.pyplot import figure

ax = log_df.plot(x='time',figsize=(12, 6), title="Number of articles extracted")
fig = ax.get_figure()
fig.savefig('Poster/log_extraction.png', transparent=True, bbox_inches='tight')
```
